```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = 'C:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... Sorry~~~')
    


tp_for_map = pd.read_csv("C://Users//whs38//캡스톤//데이터//평균강수량정규화.csv",
                         encoding='cp949')
tp_for_map.head()

tp_for_map = tp_for_map.set_index('district')
tp_for_map.info()


# heatmap
plt.figure(figsize=(12,12))
sns.heatmap(tp_for_map, annot=True, fmt='.4f', linewidths=.2, cmap='YlOrRd')
plt.title("제주도 지역별 강수량 히트맵", 
          fontdict = { 'fontsize' : 15, 'fontweight' : 'bold'}, pad = 20)
plt.savefig("C://Users//whs38//캡스톤//평균강수량정규화.png")
plt.show()
```
