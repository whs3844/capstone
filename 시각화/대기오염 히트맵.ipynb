```python
import seaborn as sns
import pandas as pd


  # matplotlib 폰트를 한글로 변경하기
import platform
import matplotlib.pyplot as plt
    
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
    
if platform.system() == 'Darwin':
        rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')


dust = pd.read_csv('C://Users//whs38//캡스톤//데이터//대기오염정규화.csv',encoding='euc-kr')
dust.head()
dust = dust.set_index("district")

plt.figure(figsize=(10,10))
sns.heatmap(dust, annot=True, fmt=".2f", cmap='Greens') # heatmap 에 숫자 기입
plt.title("제주 측정장소별 대기오염도", fontdict = { 'fontsize' : 15, 'fontweight' : 'bold'}, pad = 20)
plt.show()



plt.figure(figsize=(10,10))
sns.heatmap(dust, cmap='Greens')
plt.title("제주 측정장소별  대기오염도", fontdict = { 'fontsize' : 15, 'fontweight' : 'bold'}, pad = 20)
plt.show()
```
